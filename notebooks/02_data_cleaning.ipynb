# Data Cleaning & Data Quality Preparation

This notebook performs data cleaning before validation and database loading.

### Objectives

- Remove duplicate records
- Handle missing values
- Standardize company identifiers
- Clean text columns
- Convert numeric columns
- Save cleaned datasets

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
ROOT = Path("../")

PROCESSED = ROOT / "data" / "processed"
```

```python
companies = pd.read_csv(PROCESSED / "companies_clean.csv")
profit = pd.read_csv(PROCESSED / "profitandloss_clean.csv")
balance = pd.read_csv(PROCESSED / "balancesheet_clean.csv")
cashflow = pd.read_csv(PROCESSED / "cashflow_clean.csv")
analysis = pd.read_csv(PROCESSED / "analysis_clean.csv")
ratios = pd.read_csv(PROCESSED / "financial_ratios_clean.csv")
```

```python
datasets = {
    "Companies": companies,
    "Profit & Loss": profit,
    "Balance Sheet": balance,
    "Cash Flow": cashflow,
    "Analysis": analysis,
    "Financial Ratios": ratios
}

for name, df in datasets.items():
    print(f"{name}: {df.shape}")
```

```python
for name, df in datasets.items():
    print("\n" + "=" * 50)
    print(name)
    print(df.isnull().sum())
```

```python
for name, df in datasets.items():

    print("\n" + "="*70)
    print(name)

    missing = df[df.isnull().any(axis=1)]

    print("Rows with Missing Values :", len(missing))

    display(missing.head())
```

```python
for name, df in datasets.items():

    print("\n" + "=" * 70)
    print(name)

    print("Duplicate Rows :", df.duplicated().sum())
```

```python
tables = {
    "Profit & Loss": profit,
    "Balance Sheet": balance,
    "Cash Flow": cashflow,
    "Financial Ratios": ratios
}

for name, df in tables.items():

    print("\n" + "=" * 70)
    print(name)

    dup = df[df.duplicated(subset=["company_id", "year"], keep=False)]

    print("Duplicate (company_id, year):", len(dup))

    if not dup.empty:
        display(dup.head(10))
```
