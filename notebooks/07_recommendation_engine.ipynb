# Recommendation Engine

This notebook generates investment recommendations based on:

- Performance Score
- Growth Score
- Risk Score
- Overall Company Score
- Investment Category

The output is a recommendation dataset that classifies companies into:

- BUY
- HOLD
- SELL

```python
import pandas as pd
from pathlib import Path
```

```python
ROOT = Path("../")

OUTPUT = ROOT / "outputs"

analytics = pd.read_csv(
    OUTPUT / "final_analytics.csv"
)

analytics.head()
```

```python
print(analytics.shape)

analytics.info()
```

## Investment Recommendation Logic

Recommendations are generated using:

- Overall Score
- Risk Score
- Investment Category

Rules:

BUY  → High Overall Score + Low Risk

HOLD → Medium Overall Score

SELL → High Risk OR Poor Overall Score

```python
def recommend(row):

    if (
        row["overall_score"] >= 100
        and row["risk_score"] < 2000
    ):
        return "BUY"

    elif (
        row["overall_score"] >= 40
        and row["risk_score"] < 5000
    ):
        return "HOLD"

    else:
        return "SELL"


analytics["recommendation"] = analytics.apply(
    recommend,
    axis=1
)

analytics.head()
```

```python
analytics["recommendation"].value_counts()
```

```python
buy = analytics[
    analytics["recommendation"] == "BUY"
].copy()

buy.head()
```

```python
buy.to_csv(
    OUTPUT / "buy_recommendations.csv",
    index=False
)

print("BUY recommendations saved!")
```

```python
hold = analytics[
    analytics["recommendation"] == "HOLD"
].copy()

hold.head()
```

```python
hold.to_csv(
    OUTPUT / "hold_recommendations.csv",
    index=False
)

print("HOLD recommendations saved!")
```

```python
sell = analytics[
    analytics["recommendation"] == "SELL"
].copy()

sell.head()
```

```python
sell.to_csv(
    OUTPUT / "sell_recommendations.csv",
    index=False
)

print("SELL recommendations saved!")
```

```python
top10 = analytics.sort_values(
    by="overall_score",
    ascending=False
).head(10)

top10
```

```python
top10.to_csv(
    OUTPUT / "top10_recommendations.csv",
    index=False
)

print("Top 10 Recommendations Saved!")
```

```python
high_risk = analytics[
    analytics["risk_score"] >= 5000
].copy()

high_risk
```

```python
high_risk.to_csv(
    OUTPUT / "high_risk_alerts.csv",
    index=False
)

print("High Risk Alerts Saved!")
```

```python
recommendation_dataset = analytics.copy()

recommendation_dataset.to_csv(
    OUTPUT / "recommendation_dataset.csv",
    index=False
)

print("Recommendation Dataset Saved Successfully!")
```
