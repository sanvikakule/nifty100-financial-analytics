# Financial Performance Metrics

This notebook analyzes financial performance using the SQLite database.

## Objectives

- Connect to SQLite database
- Load financial tables
- Calculate ROE
- Calculate Operating Profit Margin
- Calculate Net Profit Margin
- Analyze EPS
- Generate Company Scorecard
- Export results for dashboard

```python
import sqlite3
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
ROOT = Path("../")

DATABASE = ROOT / "data" / "db" / "nifty100.db"

conn = sqlite3.connect(DATABASE)

print("Connected Successfully!")
```

```python
companies = pd.read_sql(
    "SELECT * FROM companies",
    conn
)

profit = pd.read_sql(
    "SELECT * FROM profitandloss",
    conn
)

balance = pd.read_sql(
    "SELECT * FROM balancesheet",
    conn
)

ratios = pd.read_sql(
    "SELECT * FROM financial_ratios",
    conn
)
```

```python
print("Companies:", companies.shape)
print("Profit & Loss:", profit.shape)
print("Balance Sheet:", balance.shape)
print("Financial Ratios:", ratios.shape)
```

## Return on Equity (ROE) Analysis

Return on Equity (ROE) measures how efficiently a company generates profit using shareholders' equity.

We will:
- Extract ROE values
- Remove missing values
- Rank companies
- Export the results

```python
roe_analysis = ratios[
    [
        "company_id",
        "year",
        "return_on_equity_pct"
    ]
].copy()

roe_analysis = roe_analysis.dropna()

roe_analysis = roe_analysis.sort_values(
    by="return_on_equity_pct",
    ascending=False
)

roe_analysis.head(10)
```

```python
OUTPUT = ROOT / "outputs"
OUTPUT.mkdir(exist_ok=True)

roe_analysis.to_csv(
    OUTPUT / "roe_analysis.csv",
    index=False
)

print("ROE Analysis Saved")
```

## Operating Profit Margin (OPM)

Operating Profit Margin indicates how efficiently a company generates operating profit from its sales.

```python
opm_analysis = profit[
    [
        "company_id",
        "year",
        "sales",
        "operating_profit",
        "opm_percentage"
    ]
].copy()

opm_analysis = opm_analysis.dropna()

opm_analysis = opm_analysis.sort_values(
    by="opm_percentage",
    ascending=False
)

opm_analysis.head(10)
```

```python
opm_analysis.to_csv(
    OUTPUT / "opm_analysis.csv",
    index=False
)

print("OPM Analysis Saved")
```

## Net Profit Margin

Net Profit Margin represents the percentage of profit generated from total sales.

```python
net_margin = profit[
    [
        "company_id",
        "year",
        "sales",
        "net_profit"
    ]
].copy()

net_margin = net_margin.dropna()

net_margin["net_profit_margin"] = (
    net_margin["net_profit"] /
    net_margin["sales"]
) * 100

net_margin = net_margin.sort_values(
    by="net_profit_margin",
    ascending=False
)

net_margin.head(10)
```

```python
net_margin.to_csv(
    OUTPUT / "net_profit_margin.csv",
    index=False
)

print("Net Profit Margin Saved")
```

## Earnings Per Share (EPS)

EPS is used to evaluate the profitability of a company on a per-share basis.

```python
eps_analysis = profit[
    [
        "company_id",
        "year",
        "eps"
    ]
].copy()

eps_analysis = eps_analysis.dropna()

eps_analysis = eps_analysis.sort_values(
    by="eps",
    ascending=False
)

eps_analysis.head(10)
```

```python
eps_analysis.to_csv(
    OUTPUT / "eps_analysis.csv",
    index=False
)

print("EPS Analysis Saved")
```

```python
scorecard = (
    roe_analysis.merge(
        eps_analysis,
        on=["company_id", "year"],
        how="inner"
    )
    .merge(
        opm_analysis[
            ["company_id", "year", "opm_percentage"]
        ],
        on=["company_id", "year"],
        how="inner"
    )
)

scorecard = scorecard.rename(
    columns={
        "return_on_equity_pct": "ROE",
        "eps": "EPS",
        "opm_percentage": "OPM"
    }
)

scorecard["Performance_Score"] = (
    scorecard["ROE"] * 0.4 +
    scorecard["EPS"] * 0.3 +
    scorecard["OPM"] * 0.3
)

scorecard = scorecard.sort_values(
    by="Performance_Score",
    ascending=False
)

scorecard.head(10)
```

```python
scorecard.to_csv(
    OUTPUT / "company_scorecard.csv",
    index=False
)

print("Company Scorecard Saved")
```
